```python
root = '/media/leelabsg-storage1/yein/research/data/adni_region'

def check(num, cn_idx, new_idx):

    df = pd.read_csv(f'/media/leelabsg-storage1/yein/research/BAE/RegionBAE/data/tmp/ukbb_cn_region.csv')
    
    current_id = df['subjectID'][new_idx]
    ukb_id = ukb_cn['id'][cn_idx]

    # index finder
    idx = 0
    for i in df['subjectID'].index:
        if ukb_id == df['subjectID'][i]:
           idx = i 
    print("Results of index finder: ", ukb_id, "is ", f"{idx}/{cn_idx}-th in ukbb_cn_region.csv")
    idx = 0
    for i in ukb_cn['id'].index:
        if current_id == ukb_cn['id'][i]:
           idx = i 
    print("Results of index finder: ", current_id, "is ", f"{idx}-th in ukb_id['id']")
    print("Check if the index of the preprocessed data is correct: ")
    print(ukb_id==current_id)
    print(ukb_cn['age'][cn_idx]==df['age'][new_idx])

    # Visualization
    current_path = os.path.join(root, str(current_id))
    
    img_path = current_path + '/T1w_registered.nii.gz'
    mask_path = current_path + '/T1w_brain_mask_registered.nii.gz'
    img = ants.image_read(img_path)
    mask = ants.image_read(mask_path)
    explore_3D_array(img.numpy())
    explore_3D_array(mask.numpy())

    # Region visualization
    for i in range(1, 10):
        region_path = current_path + f'/region_{i}.nii.gz'
        region_mask_path = current_path + f'/region_{i}_mask.nii.gz'
        region = ants.image_read(region_path)
        region_mask = ants.image_read(region_mask_path)
        explore_3D_array(region.numpy())
        explore_3D_array(region_mask.numpy()) 
```

# 현재 전처리된 데이터 id 매칭

```python
current_files = glob.glob(root + '/*')
curr_subjects = []
for i in range(len(current_files)):
    subject = current_files[i].split('/')[-1]
    curr_subjects.append(int(subject))
print(len(curr_subjects))
```

```python
total_subjects = ukb_cn['id'].tolist()
```

```python
miss_matched = []
missed = []
for i in range(len(curr_subjects)):
    if total_subjects[i] != curr_subjects[i]:
        if total_subjects[i] not in set(curr_subjects):
            missed.append(total_subjects[i])
        else:
            miss_matched.append(total_subjects[i])
```

```python
curr_set = set(curr_subjects)
total_set = set(total_subjects)
diff = total_set - curr_set
print(len(diff))
```

```python
print(len(missed), len(miss_matched))
```
