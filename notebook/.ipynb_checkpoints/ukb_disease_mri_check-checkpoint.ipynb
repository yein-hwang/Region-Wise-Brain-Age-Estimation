# UKBB icd10 Code

```python
icd101 = pd.read_csv('/media/leelabsg-storage1/DATA/icd/icd10_41202.csv')
icd102 = pd.read_csv('/media/leelabsg-storage1/DATA/icd/icd10_41270.csv')
icd91 = pd.read_csv('/media/leelabsg-storage1/DATA/icd/icd9_41203_with_icd10.csv')
icd92 = pd.read_csv('/media/leelabsg-storage1/DATA/icd/icd9_41271_with_icd10.csv')
icd101 = icd101[['eid', 'icd10', 'date']]
icd102 = icd102[['eid', 'icd10', 'date']]
icd91 = icd91[['eid', 'icd10', 'date']]
icd92 = icd92[['eid', 'icd10', 'date']]
icds = pd.concat([icd101, icd102, icd91, icd92]).drop_duplicates().reset_index(drop = True)
icds = icds.drop('date', axis=1).drop_duplicates().reset_index(drop = True)
#icds.set_index('eid').to_csv('all_icds.csv')
icds.to_csv('/media/leelabsg-storage1/yein/research/data/csv/ukbb_icd.csv', index=False)
icds
```

```python
ukb_brain_img_df = pd.read_csv('/media/leelabsg-storage1/yein/research/data/csv/ukbb_35012.csv')
ukb_brain_img_df
```

```python
ukb_brain_img_df['disease'].value_counts()
```

```python
ukb_brain_mri_icds = icds.loc[icds.eid.isin(ukb_brain_img_df['id'][ukb_brain_img_df['disease']==1])].reset_index(drop=True)
ukb_brain_mri_icds
```

### I10: 고혈압(hypertension)
### Z86.4: 과거의 정신 및 행동 장애의 병력 (Personal history of psychological and behavioral disorders)
### E780: Pure hypercholesterolemia (순수 고콜레스테롤혈증)

```python
# 질병 코드 목록
icd10_codes = ['I10', 'Z864', 'E780']

# 각 질병 코드에 대해 필터링하고 CSV로 저장
for code in icd10_codes:
    # 해당 질병 코드에 해당하는 eid 필터링
    filtered_eids = ukb_brain_mri_icds[ukb_brain_mri_icds['icd10'] == code]['eid'].unique()
    
    # 필터링된 eid를 사용하여 ukb_brain_img_df에서 일치하는 행 필터링
    filtered_df = ukb_brain_img_df[ukb_brain_img_df['id'].isin(filtered_eids)]
    
    # 필요한 컬럼 선택
    output_df = filtered_df[['id', 'Age']]
    
    # CSV 파일로 저장
    output_filename = f'/media/leelabsg-storage1/yein/research/data/csv/ukbb_{code}.csv'
    output_df.to_csv(output_filename, index=False)
```
